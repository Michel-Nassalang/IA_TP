```python
def somme():
    n = input("Donner la valeur de n :");
    n = int(n);
    somme = 0;
    for i in range(n):
        somme = somme + (i+3);
    return somme;
```

```python
s = somme();
print("La somme est de : ",s);
```

```python
def produit():
    n = input("Donner la valeur de n :");
    n = int(n);
    prod = 1;
    for i in range(n):
        prod = prod * (i+3);
    return prod;
```

```python
p = produit();
print("Le produit est de : ",p);
```
