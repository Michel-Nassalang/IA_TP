#### Analyse du graphe obtenu
On a un graphe donnant toutes les valeurs prises par les variables du dataset. On voit que la variable 'Patient age quantile'

Ce qui peut se traduire par une manque importante de d'informations pour ces variables du dataset. C'est de ce fait que nous allons évaluer l'existence des données manquantes et les traiter.

#### Pourcentage des valeurs manquantes sur le dataset

```python
valnull = data.isnull().sum().sum()
print('Nombre d\'élements nuls: ',valnull)
vals = data.shape[0]*data.shape[1]
print('Nombre d\'élements du dataset: ',vals)
p = (valnull/vals)*100
print('Pourcentage des valeurs manquantes: {}%'.format(p))
```

#### Elimination des variables avec un taux de valeurs manquantes supérieur à 90%

```python
data.isnull().sum()/data.shape[0]*100
```

```python
valaccept = (data.isnull().sum()/data.shape[0]*100)[(data.isnull().sum()/data.shape[0]*100>90).values == False]
valaccept
# valaccept est la série contenant toutes les colonnes ayant un taux de valeurs manquantes inférieur à 90%
```

```python
data_clean = data[valaccept.index]
data_clean
# data_clean est le dataframe ayant subi le nettoyage des données pour un taux de données manquantes supérieur à 90%.
# Après élimination, on a un 39 colonnes.
```

```python
data_clean = data_clean.drop(columns=['Patient ID'])
```

```python
data_clean
```
