#### Pourcentage des valeurs manquantes sur le dataset

```python
valnull = data.isnull().sum().sum()
print('Nombre d\'Ã©lements nuls: ',valnull)
vals = data.shape[0]*data.shape[1]
print('Nombre d\'Ã©lements du dataset: ',vals)
p = (valnull/vals)*100
print('Pourcentage des valeurs manquantes: {}%'.format(p))
```

#### Elimination des variables avec un taux de valeurs manquantes supÃ©rieur Ã  90%

```python
data.isnull().sum()/data.shape[0]*100
```

```python
valaccept = (data.isnull().sum()/data.shape[0]*100)[(data.isnull().sum()/data.shape[0]*100>90).values == False]
valaccept
# valaccept est la sÃ©rie contenant toutes les colonnes ayant un taux de valeurs manquantes infÃ©rieur Ã  90%
```

```python
data_clean = data[valaccept.index]
data_clean
# data_clean est le dataframe ayant subi le nettoyage des donnÃ©es pour un taux de donnÃ©es manquantes supÃ©rieur Ã  90%.
# AprÃ¨s Ã©limination, on a un 39 colonnes.
```

```python
data_clean = data_clean.drop(columns=['Patient ID'])
```

```python
data_clean
```
