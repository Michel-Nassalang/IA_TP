```python
data_clean = data_clean_NA_1.dropna(subset=['Result_collect']);
data_clean
```

```python
data_clean.isna().sum()
```

```python
X = data_clean.drop(columns=['Result_collect']).astype('int32')
```

```python
Y = data_clean[['Result_collect']].astype('int32')
```

# Spliting des échantillons

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.7, random_state=123)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# Entrainement du modèle

La fonction softmax : $$ y = \frac{1}{1 + e^{-x}}$$

```python
model = Sequential()
model.add(Dense(200, input_dim=11, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(1,activation='softmax'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
import time
start_time = time.time()
model.fit(X_train, Y_train, epochs=5, batch_size=10, verbose=0)
time_exec = time.time() - start_time;
print("Le temps passé à entrainer ce model pour 5 couches est égale:", time_exec)
```

```python
y_predict = pd.DataFrame(model.predict(X_test))
y_predict
```

```python
score = model.evaluate(X_test, Y_test, batch_size=10, verbose=0)
print("Le taux de réussite est de: ", score[1]*100)
```
